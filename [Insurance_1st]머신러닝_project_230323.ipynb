```python
y = data['charges']
X = data.drop('charges', axis =1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2023)
```

```python
X_train_1 =  X_train.replace({'sex':{'female':0, 'male':1}, 'smoker': {'no': 0, 'yes':1}, 'region':{'northeast':1, 'southeast':2, 'southwest':3, 'northwest':4}})
X_train_1.info()
```

```python
X_test_1 =  X_test.replace({'sex':{'female':0, 'male':1}, 'smoker': {'no': 0, 'yes':1}, 'region':{'northeast':1, 'southeast':2, 'southwest':3, 'northwest':4}})
X_test_1.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Train    . (train  fit   train  predict  )
reg_1 = LinearRegression()
reg_1.fit(X_train_1, y_train)

y_train_predict = reg_1.predict(X_train_1)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = round(reg_1.score(X_train_1, y_train),2)    # reg_1   .

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
X_train_2 = X_train_1[['age','bmi','smoker']]
X_train_2.head()
```

```python
# Train    . (train  fit   train  predict  )
reg_2 = LinearRegression()
reg_2.fit(X_train_2, y_train)

y_train_predict = reg_2.predict(X_train_2)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = round(reg_2.score(X_train_2, y_train),2)    # reg_1   .

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```
