# id 주소 값 주의!

## 얕은 복사
## : 값을 변경하면 다른 변수에도 영향을 끼치도록 '참조'만 복사한 것.

```python
a = [10,9,8,7,6]
b = [5,7,8,9,10]
```

```python
print(id(a), id(b))
```

```python
b = a  # 할당하면 id가 같아짐
```

```python
print(id(a),id(b))
```

```python
b
```

```python
a.sort()
```

```python
a
```

```python
b
```

```python
a = [10,9,8,7,6]
b = [5,7,8,9,10]
print(id(a),id(b))
b = a[:]
print(id(a),id(b))
b
```

```python
a.sort()
print(a,b)
```

```python
a = [10,9,8,7,6]
b = [5,7,8,9,10]
print(id(a),id(b))
b = a.copy()
print(id(a),id(b))
b
```

```python
a.sort()
print(a,b)
```

# 깊은 복사

```python
import copy
a = [10,9,8,7,6]
b = [5,7,8,9,10]
print(id(a),id(b))
b = copy.deepcopy(a)
print(id(a),id(b))
b
```

```python
a.sort()
print(a,b)
```
