```python
# list로 받을 때
def add_list(*args):
  s = 0
  for i in args:
    for j in i:
      s += j
    return s

# 단점 : list만 들어가고 tuple이 못들어감.
```

```python
add_list([1,2,3,4,5])
```

```python
# tuple과 list를 모두 받을 수 있는 함수.

def add_all3(*args):
  temp = 0
  for i in range(len(args)):
    if type(args[i]) == list:
      for j in args[i]:
        temp += j
    else:
      temp += args[i]
  return temp
```

```python
print(add_all3(1,2,3,4))
```

```python
# 재귀적으로 하지 않은 것.

def fact(n):
  f = 1
  for i in range(1, n+1):
    f = f*i
  return f
```

```python
fact(5)
```

```python
# 재귀적으로 사용할 때, recursive
def fact2(n):
  if n <= 1: # n이 1 이하면 종료.
    return 1
  return n * fact2(n-1)
```

```python
fact2(5)
```

*주의 : map의 값을 보려면 꼭 list로 감싸야 함.

```python
items = [1,2,3,4,5]
```

```python
squared = []
for i in items:
  squared.append(i*i)

print(squared)
```

# for문 + if문
## **[**저장할 값 **for** 원소 **in** 반복 가능한 객체 **if** 조건**]**
## **[**저장할 값 **if** 조건 **else** 저장할 값 **for** 원소 **in** 반복가능한 객체]

```python
even = list()
for x in range(10,21):
  if x % 2 == 0:
    even.append(x)

print(even)
```

```python
even = [x for x in range(10,21) if x%2==0]
even
```
